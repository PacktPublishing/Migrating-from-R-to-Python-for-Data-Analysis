```python
import pandas as pd
import numpy as np 
```

```python
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Mean 
tips.groupby('sex').mean() 
```

```python
# Std 
tips.groupby('sex').std() 
```

```python
# min 
tips.groupby('sex').min() 
# max
tips.groupby('sex').max() 
# count
```

```python
# describe
tips.groupby('sex').describe() 
```

```python
tips.groupby('day').agg( ['mean' , 'min'])
```

```python
tips.groupby('day').agg( {'total_bill':'mean' , 'tip':'min'})
```

```python
tips.groupby(['sex' ,'day']).agg( {'total_bill':'mean' , 'tip':'min'})
```

```python
# Pivot tables: 
```

```python
tips.pivot_table(index = 'day'  , columns='sex'  , values='total_bill')
```
