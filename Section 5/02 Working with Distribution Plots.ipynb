```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## Dist Plot: 

```python
x = iris['sepal_length']
```

```python
sns.distplot(x)
```

## Histogram: 

```python
sns.distplot(x , kde= False)
```

```python
# Changing bins 
sns.distplot(x , kde= False  , bins= 20)
```

## Joint Plot: 

```python
x = iris['sepal_length']
y = iris['petal_length']
```

```python
sns.jointplot(x , y )
```

## Pair Plots: 

```python
sns.pairplot(iris)
```
