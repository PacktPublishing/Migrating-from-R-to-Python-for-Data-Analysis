## Pivot Table: 

```python
mydis={
    'Date':['5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017','5/3/2017'],
    'city':['new york','new york','new york','mumbai','mumbai','mumbai','beijing','beijing','beijing'],
    'temp':[65,66,68,75,78,82,80,77,79],
'humidity':[56,58,60,80,83,85,26,30,35]
}
df=pd.DataFrame(mydis)
df
```

#### Draw a pivot table with date on index and city on columns: 

```python
df.pivot(index='Date',columns='city')
```

#### Draw a pivot table with date on index and city on columns and just humidity values: 

```python
df.pivot(index='Date',columns='city' , values = ['humidity'])
```

## Merging: 

Let's say we have the datasets as given below: 

```python
# Dataset one is for the temperature for corresponding dates: 
temp1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
temp
```

```python
# Dataset 2 is for additional weather related variables like wind speed and weather type like sunny, cold or rainy: 
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)
wind_speed
```

#### Join the temp and ws datasets on the basis of dates: 

```python
df=pd.merge(temp,wind_speed,on=['date','event'])
df
```
