## Merging: 

Let's say we have the datasets as given below: 

```python
# Dataset one is for the temperature for corresponding dates: 
temp1={
      "date":['01-02-12','02-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
temp
```

```python
# Dataset 2 is for additional weather related variables like wind speed and weather type like sunny, cold or rainy: 
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)
wind_speed
```

#### Join the temp and ws datasets on the basis of date and event column: 

```python
df=pd.merge(temp,wind_speed,on=['date','event'])
df
```

#### Instead of inner join, make use of `full outer join` this time to join the temp and ws datasets on the basis of date and event column: 

```python
df=pd.merge(temp,wind_speed,on=['date','event'] , how="outer" )
df
```

#### Instead of inner join, make use of `left join` this time to join the temp and ws datasets on the basis of date and event column: 

```python
df=pd.merge(temp,wind_speed,on=['date','event'] , how="left" )
df
```
